```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.rcParams['figure.figsize'] = 30, 15
```

```python
plt.quiver(0, 0, 5, 3, scale_units='xy', angles='xy', scale=1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```python
def plot_vectors(vecs):
    for vec in vecs:
        plt.quiver(vec[0], vec[1], vec[2], vec[3], scale_units='xy', angles='xy', scale=1)
    plt.xlim(min(min(vecs)), max(max(vecs)))
    plt.ylim(min(min(vecs)), max(max(vecs)))
    plt.show()
```

```python
plot_vectors([(0, 0, 3, 1), (0, 0, -3, 4)])
```
